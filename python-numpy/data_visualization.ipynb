```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"*")
```

```python
plt.plot(x,y,'r--')
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
```
